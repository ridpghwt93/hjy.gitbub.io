```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
pd.set_option('mode.chained_assignment',  None)  # warning 제거
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask') #PLM

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df = pd.read_csv('chemistry dictionary.csv', encoding='CP949')  # 왼쪽 사이드(파일)에 업로드한 파일명과 동일해야 함

df.head()
```

```python
df = df[~df['챗봇'].isna()]

df.head()
```

```python
df.loc[0, '유저']
```

```python
model.encode(df.loc[0, '유저'])
```

```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy

df['embedding'] = df['유저'].map(lambda x: list(model.encode(x)))

df.head()
```

```python
df.to_csv('chemistry dictionary.csv', index=False)
```

```python
print("1번부터 20번 원소들의 특징과 50여 개의 분자식과 화학반응식에 대해 알려드립니다."
"그만하고 싶으시다면 채팅창에 '고마워'이라고 입력해주세요.")
while 1:
    text=input("사용자: ")
    if text=="고마워": print("챗봇: 도움이 되었으면 좋겠네요. 안녕히가세요."); break
    embedding = model.encode(text)
    df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())
    answer = df.loc[df['distance'].idxmax()]
    answer=answer["챗봇"]
    print(f"챗봇: {answer}")
```
